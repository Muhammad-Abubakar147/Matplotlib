Simple plot with tittle and x,label


```python
import matplotlib.pyplot as plt
x=["Mon","Tues","Wed","Thurs","Fri","Sat"]
y=[45,56,87,9,43,65]

plt.title("Sale of a week")
plt.xlabel("Weeks")
plt.ylabel("Sales per week")
plt.plot(x,y)
plt.show()
```
