Bar plot In Matplotlib

```python
#Bar plot
import matplotlib.pyplot as plt

Product=["Soap","Gold","Feeder","toffee","Pamper"]
Sales=[1950,1000,950,1550,760]
plt.bar(Product,Sales,color="orange",label="Product sales")
plt.legend()
plt.title("Sales of products in Feburary")
plt.xlabel("Names of Products")
plt.ylabel("Sales of products")
plt.show()
```
